```python
to_dummylist=['education','workclass','marital_status','occupation','relationship','race','sex','native_country']
```

```python
def dummy_list(df,to_dummylist):
    for dummy in to_dummylist:
        dummies=pd.get_dummies(df[dummy],prefix='dummy',dummy_na=False)
        df=df.drop(dummy,1)
        df=pd.concat([df,dummies],axis=1)
    return df
```

```python
x=dummy_list(x,to_dummylist)
```
