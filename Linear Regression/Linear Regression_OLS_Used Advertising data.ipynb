```python
inputs=dataset[['TV','radio']]
```

```python
outputs=dataset['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
XTrain,XTest,YTrain,YTest=train_test_split(inputs,outputs,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(XTrain,YTrain)
```

```python
outPredicted=lr.predict(XTest)
```

```python
list(zip(YTest,outPredicted))
```

```python
YTest
```

```python
outPredicted
```

```python
from sklearn import metrics
```

```python
metrics.mean_squared_error(YTest,outPredicted)
```
